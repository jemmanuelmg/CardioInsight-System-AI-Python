```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split

# Leer el dataset y guardarlo en un dataframe df
df = pd.read_csv("heart.csv")

# Ver los encabezados de la tabla
df.head()




# Hacer primer grafico usando como eje X la etiqueta
'''
sns.countplot(x="target", data=df, palette="bwr")
plt.show()
'''




# Ver que porcentaje de pacientes en el dataset tienen o no enfermedad
countNoDisease = len(df[df.target == 0]) # comprensin de listas
countHaveDisease = len(df[df.target == 1])
print("Percentage of Patients Haven't Heart Disease: {:.2f}%".format((countNoDisease / (len(df.target))*100)))
print("Percentage of Patients Have Heart Disease: {:.2f}%".format((countHaveDisease / (len(df.target))*100)))




countFemale = len(df[df.sex == 0])
countMale = len(df[df.sex == 1])
print("Percentage of Female Patients: {:.2f}%".format((countFemale / (len(df.sex))*100)))
print("Percentage of Male Patients: {:.2f}%".format((countMale / (len(df.sex))*100)))



################# Edad #################
pd.crosstab(df.age,df.target).plot(kind="bar",figsize=(20,6))
plt.title('Enfermedad del corazn segn la edad')
plt.xlabel('Edad')
plt.ylabel('Frecuencia de la Enfermedad')
plt.legend(["Sano", "Enfermo"])
plt.savefig('FrecuenciaSegunEdad.png')
plt.show()



################# Dolor Toracico #################
pd.crosstab(df.cp,df.target).plot(kind="bar",figsize=(15,6),color=['#11A5AA','#AA1190' ])
plt.title('Enfermedad del corazn segn dolor torcico')
plt.xlabel('Tipo de Dolor Torcico')
plt.xticks(rotation = 0)
plt.ylabel('Frecuencia de la Enfermedad')
plt.legend(["Sano", "Enfermo"])
plt.savefig('FrecuenciaSegunDolorToracico.png')
plt.show()



######## Resultados Electrocardiogaficos ##########
pd.crosstab(df.restecg,df.target).plot(kind="bar",figsize=(15,6),color=['#03506F','#F05454' ])
plt.title('Enfermedad del corazn segn resultados electrocardiogrficos')
plt.xlabel('Resultados Electrocardiogrficos')
plt.xticks(rotation = 0)
plt.ylabel('Frequencia de la Enfermedad')
plt.legend(["Sano", "Enfermo"])
plt.savefig('FrecuenciaSegunResultadoElectrocardio.png')
plt.show()



########## Pendiente Segmento ST (slope) ###########
pd.crosstab(df.slope,df.target).plot(kind="bar",figsize=(15,6),color=['#DAF7A6','#FF5733' ])
plt.title('Enfermedad del corazn segn pendiente del segmento ST en pico de ejercicio')
plt.xlabel('Pendiente del Segmento ST')
plt.xticks(rotation = 0)
plt.ylabel('Frequencia de la Enfermedad')
plt.legend(["Sano", "Enfermo"])
plt.savefig('PendienteSegmentoST.png')
plt.show()



########## Nivel de Azucar en Ayunas ###########
pd.crosstab(df.fbs,df.target).plot(kind="bar",figsize=(15,6),color=['#FFC300','#581845' ])
plt.title('Enfermedad del corazn segn nivel de azucar en ayunas')
plt.xlabel('Nivel de azucar > 120 mg/dl (1 = S; 0 = No)')
plt.xticks(rotation = 0)
plt.legend(["Sano", "Enfermo"])
plt.ylabel('Frequencia de la Enfermedad')
plt.savefig('NivelAzucarAyunas.png')
plt.show()




```
